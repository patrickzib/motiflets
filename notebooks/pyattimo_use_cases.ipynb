```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.insert(0, "../")

from motiflets.competitors import *
from motiflets.plotting import *

import pandas as pd
import seaborn as sns

import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}

import warnings
warnings.simplefilter("ignore")

def read_data(dataset, sampling_factor=30_000):
    full_path = '../datasets/original/' + dataset
    
    data = pd.read_csv(full_path, header=None).squeeze('columns')    
    print("Dataset Original Length n: ", len(data))
    
    data, factor = ml.resample(data, sampling_factor)
    print("Dataset Sampled Length n: ", len(data))

    return np.ascontiguousarray(data.values), factor
```

```python
file = "dishwasher.txt" 
ds_name = "Dishwasher"
series, factor = read_data(file)

mot = Motiflets(ds_name, series, backend="default")
fig, ax = mot.plot_dataset(max_points=30_000)

print (f"Downsampled by {factor}")
```

```python
ks = 40
length_range = np.arange(25,500,25) 
motif_length = mot.fit_motif_length(ks, length_range)
print ("Best motif length", motif_length, "seconds")
```

```python
# print(motif_length)
ks = 40
motif_length = 125
dists, candidates, elbow_points = mot.fit_k_elbow(ks, motif_length)
```
