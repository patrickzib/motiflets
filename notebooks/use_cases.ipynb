# Human Behaviour

```python
path = "../datasets/PeVAMmotif/"

def read_human_behaviour(index):
    ds_name = "Human Behavior"
    test = sio.loadmat(path + 'a1va3.mat')
    series = test["data"].T
    return ds_name, series[index,:]

ds_name, series = read_human_behaviour(11)


# print (series.shape)
mm = Motiflets(ds_name, series)

# Paper: l = 200 or 100
_ = mm.fit_k_elbow(
        k_max=10, 
        motif_length=200,
        plot_elbows=True, 
        plot_motifs_as_grid=False
)

mm.plot_motifset();
```

# Penguin

```python
path = "../datasets/PeVAMmotif/"

def read_penguin():
    ds_name = "Penguin"
    test = sio.loadmat(path + 'penguinLabel.mat')
    series = test["data"].T
    print(series.shape)
    return ds_name, series[2,:1_000_000]

ds_name, series = read_penguin()
mm = Motiflets(ds_name, series)

_ = mm.fit_k_elbow(
    k_max=10, 
    motif_length=22,
    plot_elbows=True, 
    plot_motifs_as_grid=False
)

mm.plot_motifset();
```
