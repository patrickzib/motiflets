# Human Behaviour

```python
path = "../datasets/PeVAMmotif/"

def read_human_behaviour(index):
    ds_name = "Human Behavior"
    test = sio.loadmat(path + 'a1va3.mat')
    series = test["data"].T
    return series[index,:]

series = read_human_behaviour(11)


# print (series.shape)
mm = Motiflets(ds_name, series, backend="default")
mm.plot_dataset()

# Paper: l = 200 or 100
_ = mm.fit_k_elbow(
    k_max=10, 
    motif_length=200,     
    plot_elbows=True, 
    plot_motifs_as_grid=True)

_  = mm.plot_motifset()
```
